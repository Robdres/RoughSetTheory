```python
import numpy as np
import tensorflow as tf
import cv2
import random as rd
```

```python
def getCenters(n,mR): 
    centers = []
    mR = int(mR)
    if(mR>700-mR):
        centers,mR = getCenters(n,mR/2)
        return centers,mR
    else:
        cX = rd.randint(mR,700-mR)
        cY = rd.randint(mR,700-mR)
    centers.append((cX,cY))
    it = 0
    while len(centers) < n: 
        it += 1
        cX = rd.randint(mR,700-mR)
        cY = rd.randint(mR,700-mR)
        suma = 0       
        for j in centers:
            if (np.sqrt((cX-j[0])**2 + (cY-j[1])**2) < (mR+1)*2):
                suma +=1
        if suma==0:
            centers.append((cX,cY))
            
        if it>10000:
            return getCenters(n,mR/2)
    return centers,mR

```

```python
def randomCircles(n):
    width = 700
    height = 700
    circles = 0
    
    minR = 0
    maxR = 750/(n)+10
    
    centers = getCenters(n,maxR)
    maxR = centers[1]
    minR = maxR - 20
    centers = centers[0]
    img = np.ones((width,height,3), dtype=np.uint8)*255
    for i in centers:
        img = cv2.circle(img,i,rd.randint(int(minR),int(maxR)),(255,0,0))
    return img
im = randomCircles(2) 
cv2.imshow("image", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def randomSquares(n):
    width = 700
    height = 700
    circles = 0
    
    minR = 11
    maxR = 750/(n)+10
    
    centers = getCenters(n,maxR)
    maxR = centers[1]
    minR = maxR - 20
    centers = centers[0]
    img = np.ones((width,height,3), dtype=np.uint8)*255
    for i in centers:
        r = rd.randint(int(minR),int(maxR))-10
        a = rd.randint(10,80)*np.pi/180
        iP = r*np.cos(a)
        rP = r*np.sin(a)
        img = cv2.rectangle(img,(int(i[0]-iP),int(i[1]+rP)),(int(r+i[0]),int(i[1]-r)),(0,0,0),1)
    return img
im = randomSquares(10) 
cv2.imshow("image", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def randomTriangles(n):
    width = 700
    height = 700
    circles = 0
    
    minR = 11
    maxR = 750/(n)+10
    
    centers = getCenters(n,maxR)
    maxR = centers[1]
    minR = maxR - 20
    centers = centers[0]
    img = np.ones((width,height,3), dtype=np.uint8)*255
    for i in centers:
        r = rd.randint(int(minR),int(maxR))-10
        a = rd.randint(10,110)*np.pi/180
        b = rd.randint(130,230)*np.pi/180
        c = rd.randint(250,350)*np.pi/180
        p1 =  (r*np.cos(a),r*np.sin(a))
        p2 =  (r*np.cos(b),r*np.sin(b))
        p3 =  (r*np.cos(c),r*np.sin(c))
        pts = np.array([i[0]+p1[0],i[1]+p1[1],i[0]+ p2[0], i[1]+p2[1],i[0]+p3[0],i[1]+p3[1]], np.int32)
        pts = pts.reshape((-1,1,2))
        img = cv2.polylines(img,[pts],True,(0,0,0),1)
    return img
im = randomTriangles(3) 
cv2.imshow("image", im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Crear bases de datos
it = 1
for i in range(300):
    img = randomCircles(1)
    cv2.imwrite("dataSet/1Circles/1Circle_"+str(it)+".png",img)
    img = randomSquares(1)
    cv2.imwrite("dataSet/1Rectangles/1Rectangle_"+str(it)+".png",img)
    img = randomTriangles(1)
    cv2.imwrite("dataSet/1Triangles/1Triangle_"+str(it)+".png",img)
    it += 1
img = randomCircles(1)
```
